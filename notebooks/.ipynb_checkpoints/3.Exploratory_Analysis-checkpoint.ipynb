#### 2.2.1 duration subset
In section 3.1 we've observed how the duration is right-skewed, with 95% of the durations with 30minutes. We'll select this subset before diving deeper into the analysis of trip durations.

```python
duration_subset = df[df['duration_min']<30]
```

#### 2.2.2 weekly subset
In section 4.2 we've seen that we don't have the same amount of days captured for each week in our dataset, to adjust for this, we create another subset of only weekdays in week 2-5 (week one is Wed-Fri, where Wed is 01/01).

```python
week_subset = df[(df['week']>1) & (df['day_type'] == 'weekday')]
```

In section 6.1 we've observed some 'edge cases' or outliers in the distance set, which we filter out.

```python
# develop a subset for analysis
distance_subset = df[(df['distance_km'] < 10) & (df['distance_km'] > 0.2)]
distance_pcts = np.percentile(distance_subset['distance_km'], [25,50,75])
distance_pcts
```

## 5. Breakdown weekly growth
***
From the data we've seen so far, the trip-count can often be explained by behavioral data (trips day/nigh, commuting). The strong week-on-week growth however, cannot easily be explained by this. 

We want to look at metadata (~demographics) available, to see if we can find any correlation between specific groups of users and the observed growth. The metadata available:
* user_type
* location
* rental_access_method

For each of these demographic parameters, we look at the breakdown of absolute number of trips per week and the week-on-week growth of trips.

#### 5.1 user-type

```python
g = sns.countplot(data=week_subset, x='week', hue='user_type')
    
```

```python
user_growth, user_total = weekGrowthBreakdown(week_subset, 'user_type', ['Subscriber', 'Customer'])
user_growth
```

```python
g = sns.barplot(data = user_growth[user_growth['week']>2], x = 'week', y = 'growth', hue = 'user_type')
```

> As we observe overall growth in number of trips; we also observe a strong subscriber growth! Looking at the individual growth numbers, one hypothesis is that we are looking at two processes:
> 1. shift from Customer to Subscriber. A user who was a 'casual' customer becomes a subscriber (member). This means his trips will now be logged as subscriber and not anymore as customer.
> 2. general growth in usage: more people who start as subscriber and more people who start as member.

> Having a user-id (even anonymised, for security measures even location data could be dropped for this specific case) could help shed light on the customer lifecycle.

#### 5.2 Rentall Access Method

```python
g = sns.countplot(data=week_subset, x='week', hue='rental_access_method',
                 hue_order = ['app','undefined','clipper'])
```

```python
rental_growth, rental_total = weekGrowthBreakdown(week_subset, 'rental_access_method', ['app','clipper','undefined'])
rental_growth
```

```python
g = sns.barplot(data = rental_growth[rental_growth['week']>2], x = 'week', y = 'growth', 
                hue = 'rental_access_method', hue_order = ['app','undefined','clipper'])
```

> Clear and obvious is the strong growth of 'App' as a rental method (>40% week on week), whereas 'Undefined' is dropping and Clipper is also growing. Potentially, there is a shift from trips being recorded with 'undefined' as rental method to 'App'. It could be for example that only certain version of the app (or OS) are updated to respond with this parameter. 

> Another hypothesis that could partly explain what we're observing here is that 'undefined' correlates with users being a causal customer and only member have the app. This is easy to confirm and will be done in the section directly below.

> More details on this category would be needed, however it seems to be a parameter that has just been added (it's not even in the data dictionary), hence we will not put more research into this at the current time.

```python
r1 = df.groupby(['user_type','rental_access_method']).size().reset_index()
r1.columns = ['user_type','rental_access_method','trips']
r2 = r1.pivot(index = 'rental_access_method', columns = 'user_type', values = 'trips')#)#.reset_index()
r2.columns = [i for i in (r2.columns)]

# normalise per column
r2['cust_perc'] = r2['Customer'] / r2['Customer'].sum()
r2['subs_perc'] = r2['Subscriber'] / r2['Subscriber'].sum()
r2
```

> We see here that indeed most subscribers use the app to book their trips (87%), whereas customer trips are often logged as undefined (81%). Hence, the relative stronger growth of subscribers would also explain part of the growth in app compared to undefined rental method logs.

#### 6.6 distance vs. duration
We've seen in multiple breakdown plots that duration and distance show similar behavior (as expected), so let's see if we can find the correlation.

```python
plt.figure(figsize = (7,7))

# plot data
plt.scatter(data = reg_subset, x = 'duration_min', y = 'distance_km', s = 0.2, alpha = 0.1)

# correlation coefficient
corr = np.corrcoef(reg_subset['duration_min'], reg_subset['distance_km'])[0,1]

plt.xlim(0,30)
plt.ylim(-0.1,10)

# labels
plt.xlabel('trip duration (min)', size = 12)
plt.ylabel('trip distance (km)', size = 12)
plt.title('correlation: {:.2f}'.format(corr));
```

> We can indeed see a strong correlation between trip duration & trip distance of 0.69. Outliers do have a great impact on this result (mainly distance <200m)

#### 6.6.1 regression (rough)

```python
reg_coeff = np.polyfit(reg_subset['duration_min'], reg_subset['distance_km'],1)
reg_coeff
```

```python
# velocity estimate - km/h
reg_coeff[0]*60
```

```python
plt.figure(figsize = (7,7))

# plot data
plt.scatter(data = reg_subset, x = 'duration_min', y = 'distance_km', s = 0.2, alpha = 0.1)

# correlation coefficient
corr = np.corrcoef(reg_subset['duration_min'], reg_subset['distance_km'])[0,1]

# plot regression
x_fit = np.array([0,30])
y_fit = x_fit*reg_coeff[0] + reg_coeff[1]
plt.plot(x_fit, y_fit, lw = 2, c = 'k', label = 'fit')


plt.xlim(0,30)
plt.ylim(-0.1,10)

# labels
plt.xlabel('trip duration (min)', size = 12)
plt.ylabel('trip distance (km)', size = 12)
plt.title('correlation: {:.2f}'.format(corr));
```

> Regression reveals an average velocity of ~8km/h, which seems suitable for biking in a city (stop&go, traffic). The fit does eem a bit off, but it is difficult to asses the exact distributions form this scatterplot.
