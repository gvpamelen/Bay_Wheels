## 3. Assign a city to each trip
***
As there is no overlap, we could quite simply split these out and assign one of these cities to each trip.

#### 3.1 identify divisions between the clusters

```python
plt.scatter(data = df, x = 'start_station_longitude', y = 'start_station_latitude', 
            s = 3,alpha = 0.1)

# try to find a split - longitude alone coult work
plt.axvline(-122.33, c='r')
plt.axvline(-122.1, c='r')

plt.xlabel('start_station_longitude')
plt.ylabel('start_station_latitude');
```

#### 3.2 assign each trip to a city

```python
df.info()
```

```python
# split (equivalent SQL CASE) -- mind sign: we're working with negative values..
df['SF'] = (df['start_station_longitude'] < -122.33)*1
df['SJ'] = (df['start_station_longitude'] > -122.1)*1
df['BK'] = ((df['SF']==0) & (df['SJ']==0))*1

# we need a combined column
df['loc'] = df['BK']*1 + df['SJ']*2 + df['SF']*3
df['loc'] = df['loc'].replace({1 : 'BK', 2 : 'SJ', 3 : 'SF'})
```

```python
# check of out encoding
g = sns.FacetGrid(data = df.sample(1000), hue = 'loc', height = 4, aspect = 1.5)
g.map(plt.scatter, 'start_station_longitude', 'start_station_latitude', alpha = 0.3)
plt.xlabel('start_station_longitude')
plt.ylabel('start_station_latitude')
plt.legend();
```

#### 3.3 Map visualisation
***

```python
# reload, now including loc (faster than merge of geo-dataframe)
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.start_station_longitude, df.start_station_latitude), crs = {'init': 'epsg:4326'})
gdf = gdf.to_crs(epsg=3857)
gdf.crs
```

```python
# plot again
fig, ax = plt.subplots(figsize = (10,10))
gdf.sample(2000).plot(column = 'loc', ax=ax, markersize = 3)
contextily.add_basemap(ax)
ax.set_axis_off();
```

#### 3.4 check for overlap
Overlap is defined as a user starting in cluster A (i.e. Berkeley) and ending up in cluster B (i.e. San Francisco).

Note: we don't expect to see this often given that Bay Wheels sees each location as a separate service.

```python
# copy df so we keep df clean
df_end = df.copy()
```

```python
# split (equivalent SQL CASE) -- mind sign: we're working with negative values..
df_end['SF_end'] = (df_end['end_station_longitude'] < -122.33)*1
df_end['SJ_end'] = (df_end['end_station_longitude'] > -122.1)*1
df_end['BK_end'] = ((df_end['SF_end']==0) & (df_end['SJ_end']==0))*1

# we need a combined column
df_end['loc_end'] = df_end['BK_end']*1 + df_end['SJ_end']*2 + df_end['SF_end']*3
df_end['loc_end'] = df_end['loc_end'].replace({1 : 'BK', 2 : 'SJ', 3 : 'SF'})
```

```python
# check overlap
df_end[df_end['loc'] != df_end['loc_end']]
```

> In the end though, this happens for 9 trips out of 295,836 or 0.003% of the trips. Doesn't seem neccesary to filter on it. Going forward, we assume all trips occur (end + start) in the city we found from the start-station's location.

## 4. Distance of a trip
***
As we have the start and end location (lat/lon), we can calculate the distance of a trip.

Note: 
* for plotting and to properly load the data we need crs `epsg:4326`, but for a distance in meters we need `epsg:3857`
* a geopandas dataframe typically only takes one 'geo-colunn', to compare the two, we use Point in a loop.

```python
# get list with geo-points (shapely) 
start_geometry = [Point(xy) for xy in zip(df.start_station_longitude, df.start_station_latitude)]
end_geometry = [Point(xy) for xy in zip(df.end_station_longitude, df.end_station_latitude)]

# convert to geopandas dataframes
crs = {'init': 'epsg:4326'}
gdf_start = gpd.GeoDataFrame(crs=crs, geometry=start_geometry)
gdf_end = gpd.GeoDataFrame(crs=crs, geometry=end_geometry)

# convert CRS
gdf_start_2 = gdf_start.to_crs(epsg=3857)
gdf_end_2 = gdf_end.to_crs(epsg=3857)


# get a distance array
distance = np.empty(len(gdf_start_2))
for i in range(len(gdf_start_2)):
    distance[i] = gdf_start_2.geometry.iloc[i].distance(gdf_end_2.geometry.iloc[i])
    
    
# add the distance array in m and km to the dataframe
df['distance_m'] = distance
df['distance_km'] = distance/1000
```

## 5. store data
***
We've now been able to generate some additional features:
* City, in loc (categorical) and as boolean in the columns named by the cities

We will store this in another hdf file.

```python
# convert loc to categorical
df['loc'] = df['loc'].astype('category')
df.info()
```

```python
# remove the geometry (from geopandas)
df.drop('geometry', axis = 1, inplace=True)
```
